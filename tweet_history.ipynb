```python
import ast
# Create a new DataFrame without rows where 'referenced_tweets' is None
new_df = df[df['referenced_tweets'].notna()].copy()

# Convert string representations to actual lists
new_df['referenced_tweets'] = new_df['referenced_tweets'].apply(ast.literal_eval)

# Extracting values into new columns
new_df['new_id'] = new_df['referenced_tweets'].apply(lambda x: x[0]['id'])
new_df['type'] = new_df['referenced_tweets'].apply(lambda x: x[0]['type'])

# Resetting the index after filtering
new_df = new_df.reset_index(drop=True)

# Display the new DataFrame
print(new_df)
```

```python
new_df.to_csv('output.csv', index=False)
```
